```python
class pow2:
    pow2 = 1

    def __init__(self, nombre_de_funcion='calculo', dato=3):
        self.nombre_de_funcion = nombre_de_funcion
        self.dato = dato

    def calculo_cuadrado(self, dato):
        self.dato = dato ** 2

    def calculo_cubo(self, dato):
        self.dato = dato ** 3

    def calculo_mod2(self, dato):
        self.dato = dato % 2


tarea = pow2(nombre_de_funcion='matematica', dato= 3)
tarea.calculo_cubo(3)
print(tarea.dato)
```
